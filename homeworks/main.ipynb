```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df.head()
```

```python
df['target'] = data.target
df['diagnosis'] = df['target'].map({0: 'Malignant', 1: 'Benign'})
df
```

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x='diagnosis')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(16, 16))
correlation_matrix = df.iloc[:, :-2].corr()
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, square=True, cbar_kws={"shrink": .8}, mask=mask)
plt.show()
```

```python
selected_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area']
```

```python
fig, axis = plt.subplots(2, 2, figsize=(15, 10))
axis = axis.ravel()
for i, feature in enumerate(selected_features):
    df[df['target'] == 0][feature].hist(bins=30, alpha=0.6, label='Malignant', ax=axis[i])
    df[df['target'] == 1][feature].hist(bins=30, alpha=0.6, label='Benign', ax=axis[i])
    axis[i].set_title(f'Распределение {feature}')
    axis[i].set_xlabel(feature)
    axis[i].set_ylabel('Частота')
    axis[i].legend()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()

for i, feature in enumerate(selected_features):
    sns.boxplot(x='diagnosis', y=feature, data=df, ax=axes[i])
    axes[i].set_title(f'Распределение {feature} по диагнозам')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='mean radius', y='mean texture', hue='diagnosis',
                data=df, palette='viridis', s=60, alpha=0.8)
plt.title('Взаимосвязь между средним радиусом и текстурой')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.legend(title='Diagnosis')
plt.show()
```
